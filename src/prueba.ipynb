```python
# Filtramos solo los Headlines con valores validos
df_interim = df_interim[df['Headline'].notna()]
df_interim.info()
```

```python
df_interim = df_interim.drop(columns=[
    'Imported Profiles',
    'Title - Publications', 
    'Company Name - Volunteering', 
    'Enrich LI Profile', 'First Name', 
    'Last Name', 'Enrich Person', 
    'Company Domain', 
    'Course Number - Courses', 
    'Company Name - Certifications',
    ])
df_interim.head()
```

```python
df_int_1['Location'].sample(10)
```

```python
# Función para extraer el país y la ciudad
def extract_location(location):
    if isinstance(location, str):  # Verifica si es una cadena
        parts = location.split(',')
        country = parts[-1].strip()  # Última parte es el país
        city = ','.join(part.strip() for part in parts[:-1]).strip()  # Resto es la ciudad
        return pd.Series([country, city])
    else:
        return pd.Series([None, None])  # Devuelve None si no es una cadena

# Aplicar la función y crear nuevas columnas
df_int_1[['Country', 'City']] = df_int_1['Location'].apply(extract_location)

# Mostrar el DataFrame resultante
print(df_int_1)
```
